```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)
```

```python
df = pd.DataFrame(X,columns=['X1','X2'])
```

```python
plt.scatter('X1', 'X2', data=df)
plt.show()
```

Com os dados gerados, podemos agora aplicar o DBSCAN para encontrar clusters


```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
clusters = dbscan.fit_predict(df)
df['clusters'] = clusters
```

```python
plt.scatter('X1', 'X2', data=df, c='clusters', cmap='rainbow')
plt.show()
```

 Atribuímos a cada ponto um rótulo correspondente ao seu cluster (vermelho, amarelo, verde e azul). Consideramos pontos com rótulo -1 como ruídos (lilás)
